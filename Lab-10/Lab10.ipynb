# Step 1: Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons,make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import NearestNeighbors
from scipy.cluster.hierarchy import linkage,dendrogram
```

# Step 2: Generate dummy data using make_blobs

```python
x,y = make_blobs(n_samples=300, random_state=42)
x = StandardScaler().fit_transform(x)
```

```python
x
```

# Step 3: Plot Generated data 

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],s=5,cmap='virdis',)
plt.title("Raw data")
plt.show()
```

# Step 4: Apply Hierarchical Clustering

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model = AgglomerativeClustering(n_clusters=3)
```

```python
model.fit(x)
```

# Step 5: Display Dendrogram

```python
linked = linkage(x,method='ward')
plt.figure(figsize=(10,5))
dendrogram(linked)
plt.show()
```

# Step 6: Generate dummy data using make_moons

```python
x,y = make_moons(n_samples=100, noise=0.1,random_state=50)
x = StandardScaler().fit_transform(x)
```

# Step 7: Plot Generated data

```python
plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],s=30,cmap='virdis')
plt.title("Raw Data")
plt.show()
```

# Step 8: apply DBSCAN Clustering 

```python
dbscan = DBSCAN(eps=0.2, min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=dbscan.labels_,cmap='rainbow',edgecolors='k')
```

# Step 9: apply K Mean Clustering on same data

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans.fit_predict(x)

plt.figure(figsize=(6,6))
plt.scatter(x[:,0],x[:,1],c=kmeans_labels,cmap='viridis',s=30)
plt.title("K-Means Clustering Result")
plt.show()
```

# Mall Customers Segmentation using DBSCAN

### Perform Following steps on Dataset for  customer segmentation
-> Load & Explore the Dataset <br>
-> Data Preprocessing <br>
-> Apply DBSCAN <br>
-> Visualizing Customer Segments <br>
-> Outliers (high-spending VIP customers or unusual shoppers <br>

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df
```

```python
##print(df.info())
##print(df.describe())
##print(df.isnull().sum())
data = df[['Annual Income (k$)', 'Spending Score (1-100)']]
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
```

```python
##dbscan = DBSCAN(eps=0.2, min_samples=5)
##dbscan.fit(x)
dbscan = DBSCAN(eps=0.5, min_samples=5)
df['Cluster'] = dbscan.fit_predict(data_scaled)
```

```python
dbscan.labels_
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Annual Income (k$)'], y=df['Spending Score (1-100)'], hue=df['Cluster'], palette='viridis')
plt.title('DBSCAN Customer Segmentation')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
outliers = df[df['Cluster'] == -1]
print("Outliers (Potential VIP Customers or Unusual Shoppers):")
print(outliers)
```
