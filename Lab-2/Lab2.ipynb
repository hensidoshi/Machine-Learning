## Step 1. Import the necessary libraries

```python
import pandas as pd
```

## Step 2. Import the dataset

```python
iphones = pd.read_csv("iphones.csv")
display(iphones)
```

## Step 3. Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 4. Create LinearRegression Model

```python
lr = LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
x = iphones['iphonenumber'].values.reshape(-1,1)
y = iphones['price']
```

## Step 6. Fit the Model (Training)

```python
lr.fit(x,y)
```

## Step 7. Prdict the value

```python
lr.predict([[14]])
predicated = lr.predict(x)
```

## Step 8. Result Visualization

```python
import matplotlib.pyplot as plt

plt.plot(iphones['iphonenumber'],iphones['price'],c='r',marker="*")
plt.plot(iphones['iphonenumber'],predicated,c='b',marker="^")
```
